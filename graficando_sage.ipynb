```python
from sage.all import EllipticCurve, gcd, QQ

def is_congruent_number(n):
    # Define the elliptic curve
    E = EllipticCurve([-n^2, 0])

    # Check if the elliptic curve has positive rank
    rank = E.rank()

    # If the rank is positive, it's a congruent number
    return rank > 0

def find_right_triangle_sides(n):
    # Check if n is a congruent number
    if not is_congruent_number(n):
        return None

    # Define the elliptic curve
    E = EllipticCurve([-n^2, 0])

    # Find a point on the curve with rational coordinates
    P = E.gens()[0]

    # Compute the coordinates of the point
    x, y = P.xy()

    # Find rational numbers m and n such that x = m/n
    m, n = QQ(x).numerator(), QQ(x).denominator()

    # Find rational numbers p and q such that y = p/q
    p, q = QQ(y).numerator(), QQ(y).denominator()

    # Calculate the sides of the right triangle
    a = n^2 - m^2
    b = 2 * m * n
    c = n^2 + m^2

    # Ensure that the sides are in reduced form
    common_divisor = gcd([a, b, c])
    a, b, c = a // common_divisor, b // common_divisor, c // common_divisor

    return a, b, c

# Example: Check if 6 is a congruent number and find the sides of the right triangle
number_to_check = 5
if is_congruent_number(number_to_check):
    sides = find_right_triangle_sides(number_to_check)
    print(f"The number {number_to_check} is congruent.")
    print(f"Sides of the right triangle: {sides}")
else:
    print(f"The number {number_to_check} is not congruent.")

    
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def boy_surface(u, v):
    x = np.sin(u) * np.sqrt(2 * np.cos(2 * u)) * np.sin(2 * v) / (np.sqrt(2) + np.sin(u) ** 2 * np.sin(2 * v))
    y = np.sin(u) * np.sqrt(2 * np.cos(2 * u)) * np.cos(2 * v) / (np.sqrt(2) + np.sin(u) ** 2 * np.sin(2 * v))
    z = np.sqrt(2) * np.sin(u) * np.cos(u) / (np.sqrt(2) + np.sin(u) ** 2 * np.sin(2 * v))
    return x, y, z

u = np.linspace(0, 2 * np.pi, 300)
v = np.linspace(0, np.pi, 150)

u, v = np.meshgrid(u, v)

x, y, z = boy_surface(u, v)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis', rstride=5, cstride=5, linewidth=0, antialiased=True)

plt.show()
```

```python
def digito_rif(ci):
    '''
    Toma un número de cédula o RIF y verifica
    el dígito validador.
    Ejemplo: V08446211 o J12345678
    '''
    base = {'V': 1 * 4, 'E': 2 * 4, 'J': 3 * 4}
    oper = [0, 3, 2, 7, 6, 5, 4, 3, 2]
    for i in range(len(ci[:9])):
        if i == 0:
            val = base.get(ci[0], 0)
        else:
            val += oper[i] * int(ci[i])
    digit = 11 - (val % 11)
    return '%s%s' % (ci[:9], str(digit)[-1])

# Ejemplo de uso
resultado = digito_rif('V08446211')
print(resultado)

```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sage.all import *

# Definir la función paramétrica
def boy_surface(u, v):
    w = CDF(u * exp(I * v))
    g1 = CDF(-3/2 * ((w - w**5) / (w**6 + sqrt(5) * w**3 - 1))).imag()
    g2 = CDF(-3/2 * ((w + w**5) / (w**6 + sqrt(5) * w**3 - 1))).real()
    g3 = CDF((1 + w**6) / (w**6 + sqrt(5) * w**3 - 1)).imag() - 1/2
    nrm = g1**2 + g2**2 + g3**2
    return vector([g1/nrm, g2/nrm, g3/nrm])

# Crear una malla de puntos (u, v)
u_values = srange(0, 2*pi, 0.1)
v_values = srange(0, 2*pi, 0.1)
points = [(u, v) for u in u_values for v in v_values]

# Evaluar la función paramétrica en cada punto de la malla
surface_points = [(u, v, boy_surface(u, v)) for (u, v) in points]

# Obtener las coordenadas para plotear
x_values = [point[2][0] for point in surface_points]
y_values = [point[2][1] for point in surface_points]
z_values = [point[2][2] for point in surface_points]

# Crear el gráfico de superficie con matplotlib
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_trisurf(x_values, y_values, z_values, color="cyan", alpha=0.7)

# Mostrar el gráfico
plt.show()
```

```python
def base10_to_base60(number):
    if number == 0:
        return "0"

    base60_digits = []
    while number:
        number, remainder = divmod(number, 60)
        base60_digits.append(str(remainder))

    return "".join(reversed(base60_digits))

# Ejemplo con el número 14646
number_in_base60 = base10_to_base60(14646)
print(number_in_base60)

```

```python
#la representación de un campo finito en forma de toroide 3D se refiere a una técnica 
#de visualización que permite observar la estructura cíclica de los elementos en un campo 
# finito de manera más intuitiva. Cada punto en el toroide representa un elemento único del campo finito, 
#y la posición del punto en el toroide está determinada por las coordenadas del elemento en el campo.

#La representación del toroide para campos finitos es un concepto más abstracto y no necesariamente se refleja directamente en una visualización 3D simple.

import matplotlib.pyplot as plt
import numpy as np

# Definir el tamaño del campo finito
p = 7

# Crear una malla de puntos en el campo finito
x = np.arange(0, p, 1)
y = np.arange(0, p, 1)
X, Y = np.meshgrid(x, y)

# Visualizar como un toroide
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, np.zeros_like(X), cmap='viridis', edgecolor='k')

# Configurar etiquetas
ax.set_xlabel('Elementos del campo finito')
ax.set_ylabel('Elementos del campo finito')
ax.set_zlabel('0')

plt.show()

```
